## 4. Feature Engineering

```python
df = pd.read_csv('student_classification_data.csv') 
df.head()
```

```python
print(df.columns)
```

### Apply Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Get dummies for multi-class categorical variables
df = pd.get_dummies(df, columns=['ParentalEducation', 'FamilyIncome'], drop_first=True)

# Label encoding for binary variables
le = LabelEncoder()
df['InternetAccess'] = le.fit_transform(df['InternetAccess'])
df['ExtraClasses'] = le.fit_transform(df['ExtraClasses'])
```

```python
df.head()
```

## 5. Feature Selection

```python
X = df[['StudyHours', 'Attendance', 'InternetAccess', 'ExtraClasses', 'HealthStatus',
        'ParentalEducation_HighSchool', 'ParentalEducation_Masters', 'ParentalEducation_PhD',
        'FamilyIncome_Low', 'FamilyIncome_Medium']]

y = df['Pass']
```

```python
X.head()
```

```python
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Convert all columns to numeric (float)
X_num = X.copy().astype(float)

# 2. Add constant/intercept term
X_const = add_constant(X_num)

# 3. Compute VIF for each feature (including the constant)
vif_data = pd.DataFrame({
    "feature": X_const.columns,
    "VIF": [
        variance_inflation_factor(X_const.values, i)
        for i in range(X_const.shape[1])
    ]
})

# 4. Drop the 'const' row from final table (only for display)
vif_df = vif_data[vif_data['feature'] != 'const'].reset_index(drop=True)

print(vif_df)
```

```python
sns.barplot(x='VIF', y='feature', data=vif_df, palette='viridis')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of features to include
features = [
    'StudyHours', 'Attendance', 'InternetAccess', 'ExtraClasses', 'HealthStatus',
        'ParentalEducation_HighSchool', 'ParentalEducation_Masters', 'ParentalEducation_PhD',
        'FamilyIncome_Low', 'FamilyIncome_Medium', 'Pass'
]

# Subset the dataframe to include only the specified features
df_subset = df[features]

# Compute the correlation matrix for the subset
corr_matrix = df_subset.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix, annot=True,fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Features vs. Pass')
```

+1: Perfect positive correlation — as one variable increases, the other increases proportionally.

0: No linear correlation — changes in one variable do not predict changes in the other.

-1: Perfect negative correlation — as one variable increases, the other decreases proportionally.

## 6. Model Building

```python
# 6.1 Train Test Split
```

```python
X = df[['StudyHours', 'Attendance', 'InternetAccess', 'ExtraClasses', 'HealthStatus',
        'ParentalEducation_HighSchool', 'ParentalEducation_Masters', 'ParentalEducation_PhD',
        'FamilyIncome_Low', 'FamilyIncome_Medium']]

y = df['Pass']
```

```python
y.value_counts()
```

```python
y.value_counts(normalize=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2, stratify=y)
```

```python
# Why stratify=y 
# Both y_train and y_test will also have ~69.5% pass and 30.5% fail.
# So your model gets a balanced view of both classes during training and testing.

# Without stratify?
# You could randomly end up with:

# Train set having 80% pass

# Test set having 50% pass
# ❌ → This would hurt model accuracy and evaluation metrics!
```

```python
y_test.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)
```

### 7. Model Instantiation

```python
log_clf = LogisticRegression()
log_clf.fit(X_train, y_train)
```

```python
log_clf.coef_
```

## 8 Evaluation

```python
def evaluation(string, model, ind_var, act):  # TESTING DATA EVALUATION
    print(string.center(50, '*'))
    pred = model.predict(ind_var)

    acc_score = accuracy_score(act, pred)
    print('Accuracy Score :', acc_score)
    print('**' * 20)

    cnf_matrix = confusion_matrix(act, pred)
    print('Confusion Matrix :\n', cnf_matrix)
    print('**' * 20)

    clf_report = classification_report(act, pred)
    print('Classification Report :\n', clf_report)
    print('**' * 20)

    return pred

y_pred_test = evaluation('Test Data Evaluation', log_clf, X_test, y_test)
```

```python
# 0 >> neg
# 1 >> pos

# Confusion Matrix :
#  [[11  1]
#  [ 1 27]]

#              Predicted
#              0    1
# Actual  0   [TN   FP]
#         1   [FN   TP]

#              Predicted
#              0    1
# Actual  0   [11   1]
#         1   [1   27]
```

```python
# precision = TP/TP+FP
precision = 27/(27+1)         # Precision for class 1
round(precision,2)
```

```python
# precision = TN/TN+FN
precision = 11/(11+1)         # Precision for class 0
round(precision,2)
```

```python
# recall = TP/TP+FN
recall = 27/(27+1)         # Recall for class 1
round(recall,2)
```

```python
recall = 11/(11+1)          # Recall for class 0
round(recall,2)
```

```python
f1_score = 2 * (precision * recall) / (precision + recall)
```

```python
f1_score = (2 * 0.96 *0.96 )/(0.96 + 0.96)   # f1 for class 1
round(f1_score,2)
```

```python
f1_score = (2 * 0.92 * 0.92) / (0.92 + 0.92)   # F1 for class 0
print(round(f1_score, 2))
```

```python
y_pred_train = evaluation('Train Data Evaluation', log_clf, X_train, y_train)  # TRAINING DATA EVALUATION
```
